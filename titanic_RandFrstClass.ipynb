```python
import pandas as pd
import numpy as np

train_data = pd.read_csv('train.csv', index_col='PassengerId')
train_data.head()
```

```python
# how many values are NaN [%]
train_data.isna().sum()/train_data.index.size
```

```python
# dropping tables that dont seem important
tables_to_drop = ['Name', 'Ticket', 'Cabin', 'Embarked']

less_train_data = train_data.drop(tables_to_drop, axis=1)
less_train_data.head()
```

```python
# got to fill NaNs in Age column
# Age might by likely estimated based on parch
less_train_data.groupby('Parch').Age.mean()
```

```python
# filling NaNs based on means above
less_train_data.loc[less_train_data.Parch == 0, 'Age'] = 32
less_train_data.loc[less_train_data.Parch == 1, 'Age'] = 24
less_train_data.loc[less_train_data.Parch == 2, 'Age'] = 17
less_train_data.loc[less_train_data.Parch == 3, 'Age'] = 33
less_train_data.loc[less_train_data.Parch == 4, 'Age'] = 44
less_train_data.loc[less_train_data.Parch == 5, 'Age'] = 39
less_train_data.loc[less_train_data.Parch == 6, 'Age'] = 43
```

```python
# converting Sex to binary
# 1 - male
# 0 - female
males = less_train_data.Sex=='male'
females = less_train_data.Sex=='female'
less_train_data.loc[males, 'Sex'] = 1
less_train_data.loc[females, 'Sex'] = 0
less_train_data.head()
```

```python
# creating RandomForestClassifier model
from sklearn.ensemble import RandomForestClassifier

X = less_train_data.drop('Survived', axis=1).values
y = less_train_data.loc[:,'Survived'].values
clf = RandomForestClassifier(n_estimators=100, max_depth=5)
clf.fit(X,y)
```

```python
# laoding test data
test_data = pd.read_csv('test.csv', index_col='PassengerId')
test_data.head()
```

```python
# how many values are NaN [%]
test_data.isna().sum()/test_data.index.size
```

```python
# same preparation as before
less_test_data = test_data.drop(tables_to_drop, axis=1)
less_test_data.loc[less_test_data.Parch == 0, 'Age'] = 32
less_test_data.loc[less_test_data.Parch == 1, 'Age'] = 24
less_test_data.loc[less_test_data.Parch == 2, 'Age'] = 17
less_test_data.loc[less_test_data.Parch == 3, 'Age'] = 33
less_test_data.loc[less_test_data.Parch == 4, 'Age'] = 44
less_test_data.loc[less_test_data.Parch == 5, 'Age'] = 39
less_test_data.loc[less_test_data.Parch == 6, 'Age'] = 43
males = less_test_data.Sex=='male'
females = less_test_data.Sex=='female'
less_test_data.loc[males, 'Sex'] = 1
less_test_data.loc[females, 'Sex'] = 0
```

```python
# there are still 3 rows with nans
less_test_data[less_test_data.isnull().any(axis=1)]
```

```python
# its hard to estimate those values, so I will just fill them with column means
less_test_data = less_test_data.apply(lambda x: x.fillna(x.mean()),axis=0)
```

```python
prediction = clf.predict(less_test_data)
```

```python
# saving data to desired format
df = pd.DataFrame(data=prediction, index=less_test_data.index, columns=['Survived'])
df.head()
```

```python
df.to_csv('my_submission.csv', index=True)
```
